```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
with open(r"Checkout Best Food Places To Eat In Bengaluru _ Zomato.html",encoding='utf') as html:
    content = html.read()
content
```

```python
#Reading the content
soup = BeautifulSoup(content,'lxml')
```

```python
restaurant_name=soup.find_all('h4',class_='sc-1hp8d8a-0 sc-bpubUI eKeNGz')
restaurant_name
```

```python
# Converting to restaurant name
Restaurant=[]
for i in restaurant_name:
    f=i.text.strip()
    Restaurant.append(f)
Restaurant
```

```python
#Ratings
ratings=[]
ratings=soup.find_all('div',class_='sc-1q7bklc-1 cILgox')
```

```python
Ratings=[]
for i in ratings:
    f=i.text.strip()
    Ratings.append(f)
Ratings
```

```python
cusine=[]
cusine=soup.find_all('p',class_='sc-1hez2tp-0 sc-sVRsr gVSfmH')
```

```python
Cusine=[]
for i in cusine:
    f=i.text.strip()
    Cusine.append(f)
Cusine
```

```python
price=[]
price=soup.find_all('p',class_='sc-1hez2tp-0 sc-sVRsr jIrGje')
```

```python
Price=[]
for i in price:
    f=i.text.strip()
    Price.append(f)
Price
```

```python
#Removing for two in price
New_Price = [s.replace('for two', '').strip() for s in Price]
#Removing ₹
New_Price=[s.replace('₹', '').strip() for s in New_Price]
# Removing comma
New_Price=[s.replace(',', '').strip() for s in New_Price]
# Converting to int
New_Price = [int(x) for x in New_Price]
# Price of 1 so diving by 2
New_Price= [x/2 for x in New_Price]
New_Price
```

```python
links=soup.find_all('a',class_='sc-gLdKKF kSLcCi')
#Gettings links
Links=[]
for i in links:
    e=i['href']
    Links.append(e)

Links
```

```python
# Offers
offers=soup.find_all('div',class_='walkin-offer-value')
offers
```

```python
#distance
distance=soup.find_all('div',class_='min-basic-info-right')
distance
```

```python
Distance=[]
for d in distance:
    k=d.text.strip()
    Distance.append(k)

Distance
```

```python
x=soup.find_all('div',class_='sc-kDgGX fjYwDL')
```

```python
new=[]
for i in x:
    new.append(i.text)
print(new)
```

```python
#Extracting discount if not there putting -
import regex as re
def extract_flat_off(s):
    match = re.search(r'\nFlat.*?OFF', s)
    return match.group(0) if match else '-'
```

```python
new = [s.replace('Flat', '\nFlat') for s in new]
new
```

```python
new_list = [extract_flat_off(s) for s in new]
new_list
```

```python
import re

def extract_discount(s):
    match = re.search(r'(\d+% OFF)', s)
    return match.group(1) if match else '-'
```

```python
discounts = [extract_discount(s) for s in new_list]
print(discounts)
```

```python
offers = discounts
```

```python
data = {
    'Restaurant': Restaurant,
    'Ratings': Ratings,
    'Cusine': Cusine,
    'Price_For_1': New_Price,
    'Links': Links,
    'Distance': Distance,
    'Offers':offers
}
```

```python
# Making dataframe from all the lists
df = pd.DataFrame(data)
df
```

```python
df.to_csv('Zomato.csv', index=False)
```

```python
# Part-2
with open(r"Reviews of High Ultra Lounge - Sheraton Grand Bangalore Hotel, Malleshwaram, Bangalore _ Zomato.html",encoding='utf') as html:
    content = html.read()
```

```python
#Reading the content
soup = BeautifulSoup(content,'lxml')
```

```python
Name=soup.find_all('p',class_='sc-1hez2tp-0 sc-gVZiCL jPENXt')
Name
```

```python
Reviews=soup.find_all('span',class_='sc-ljUfdc NCZxw')
Followers=soup.find_all('span',class_='sc-ljUfdc sc-cgThhu gQgAah')
Stars=soup.find_all('div',class_='sc-1q7bklc-1 cILgox')
Type=soup.find_all('div',class_='sc-1q7bklc-9 dYrjiw')
Time=soup.find_all('p',class_='sc-1hez2tp-0 fKvqMN time-stamp')
Text=soup.find_all('p',class_='sc-1hez2tp-0 sc-fxMfqs jBfxuu')
votes=soup.find_all('p',class_='sc-1hez2tp-0 fKvqMN')
```

```python
Names=[]
for i in Name:
    f=i.text.strip()
    Names.append(f)

Review=[]
for i in Reviews:
    f=i.text.strip()
    Review.append(f)

followers=[]
for i in Followers:
    f=i.text.strip()
    followers.append(f)

stars=[]
for i in Stars:
    f=i.text.strip()
    stars.append(f)
stars=stars[2:7]

type_of_rev=[]
for i in Type:
    f=i.text.strip()
    type_of_rev.append(f)

time=[]
for i in Time:
    f=i.text.strip()
    time.append(f)

text=[]
for i in Text:
    f=i.text.strip()
    text.append(f)


Votes=[]
for i in votes:
    f=i.text.strip()
    Votes.append(f)
```

```python
data = {
    'Names': Names,
    'Reviews': Review,
    'Followers': followers,
    'Stars': stars,
    'Type': type_of_rev,
    'Time': time,
    'Text':text,
    'Votes':Votes
}
```

```python
df=pd.DataFrame(data)
df
```

```python
df.to_csv('Zomato_Reviews.csv')
```
